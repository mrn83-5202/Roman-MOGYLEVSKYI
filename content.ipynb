### Тема 1. Заняття 3. Основи створення інтегрованого середовища розробки засобів ІАЗ.

**1. Основи використання системи контролю версій засобів ІАЗ.**

*   **Короткий опис:** Військові проекти ІАЗ, як і будь-які інші програмні проекти, потребують ефективного управління версіями коду та даних. Системи контролю версій (СКВ), такі як Git, дозволяють відстежувати зміни, співпрацювати в команді, відновлювати попередні версії та забезпечувати цілісність проекту. На цьому занятті слухачі отримають базові знання та навички роботи з Git, необхідні для ефективного управління версіями у своїх проектах ІАЗ, з акцентом на використання VScode як основного інструменту.

*   **Завдання для самостійного виконання:**
    *   Встановити Git на свій комп'ютер та інтегрувати його з VScode.
    *   Створити локальний репозиторій Git для свого проекту ІАЗ у VScode.
    *   Додати файли до репозиторію та зробити перший коміт, використовуючи інтерфейс VScode.
    *   Створити гілку для нової функціональності та внести зміни до коду.
    *   Злити зміни з гілки в основну гілку, використовуючи інструменти VScode для вирішення конфліктів злиття (якщо вони виникнуть).
    *   Переглянути історію комітів у VScode та відновити попередню версію файлу.

**2. Основи використання інтегрованого середовища розробника VScode для засобів ІАЗ.**

*   **Короткий опис:** Visual Studio Code (VScode) - це потужне та гнучке інтегроване середовище розробки, що підтримує безліч мов програмування та має широкий набір розширень для різних завдань. На цьому занятті слухачі ознайомляться з основними можливостями VScode та навчаться використовувати його для розробки своїх проектів ІАЗ, включаючи роботу з Python, Postgres та іншими інструментами.

*   **Завдання для самостійного виконання:**
    *   Встановити та налаштувати VScode на своєму комп'ютері.
    *   Встановити необхідні розширення для роботи з Python та Postgres.
    *   Відкрити свій проект ІАЗ в VScode.
    *   Використати функції VScode для написання, налагодження та запуску коду Python.
    *   Підключитися до бази даних Postgres з VScode та виконати прості запити.
    *   Використати інтегровану систему контролю версій Git для управління версіями проекту.

**Загальні рекомендації:**

*   Завдання повинні бути простими та зрозумілими для слухачів з початковим рівнем ІТ-знань.
*   Надайте докладні інструкції та приклади для кожного завдання, враховуючи використання VScode.
*   Забезпечте підтримку та допомогу під час виконання завдань.
*   Заохочуйте слухачів до експериментів та самостійного дослідження можливостей VScode та Git.

**Мета завдань:**

*   Ознайомити слухачів з основами Git та можливостями VScode.
*   Розвинути практичні навички роботи з цими інструментами в контексті розробки ІАЗ.
*   Підготувати слухачів до подальшої роботи над проектом ІАЗ, використовуючи VScode як основне середовище розробки.

**Очікуваний результат:**

*   Слухачі зможуть самостійно створювати та керувати репозиторіями Git у VScode.
*   Слухачі зможуть використовувати VScode для написання, налагодження та запуску коду Python, а також для взаємодії з базою даних Postgres.
*   Слухачі зможуть інтегрувати Git з VScode для управління версіями проекту.
